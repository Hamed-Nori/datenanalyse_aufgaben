```python
# erforderliche biblitheken importieren 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Speichere Pfaddatei von ice Cream rating.csv in einem Variable namens 'path'
path = "../../data/Ice Cream Ratings.csv"

# verbinde CSV datei mit pandas als DataFrame

df = pd.read_csv(path)
```

```python
# Gebe einen SchnellÃ¼bersicht Ã¼ber den Datenset

df
```

```python
# Benutze die Spalte Date zur indexierung von datenset
# Google Suchbegriff: pandas change index column

df = df.set_index('Date') # --> Spalte 'Date' wird nach ganz links gerÃ¼ckt
```

```python
# Wende die Methode plot aus pandas und interpretiere es--> mehr informationen dazu: 'https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html'
# output: Liniendiagramm
df.plot()
```

```python
# Obenstehende Diagramm als SÃ¤ulendiagramm erstellen
# Google Suchbegriff: how to change visual type by using plot in pandas
# Besipielsquelle: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.plot.html
df.plot(kind= "bar")
```

```python
# Erstelle mithilfe plot Methode aus Pandas seperate Liniendigaramme je Spalte
# Google Suchbegriff: How to create separate visuals by using plot in pandas
# Beiespielsquelle: https://pandas.pydata.org/docs/getting_started/intro_tutorials/04_plotting.html

df.plot(subplots=True) # --> Standard Paramter 'subplots' ist auf False gesetzt

# Gebe dein Visaul ein Title --> Ice Cream Ratings
# Setze die Beschriftung auf der X-Achse auf 'Daily Ratings'
# Setze die Beschriftung auf der Y-Achse auf 'Scores'


df.plot(subplots=True, title="Ice Cream Ratings",xlabel="Daily Ratings",ylabel="Scores" )
```
